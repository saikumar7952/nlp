```python
!pip install textblob pandas
import pandas as pd
from textblob import TextBlob
```

```python
url='/content/file.csv.zip'
data=pd.read_csv(url)
```

```python
def analyze_sentiment(text):
  analysis=TextBlob(text)
  return analysis.sentiment.polarity
```

```python
data['sentiment']=data['labels'].apply(analyze_sentiment)
```

```python
def classify_sentiment(polarity):
  if polarity>0:
    return 'positive'
  elif polarity<0:
    return 'negative'
  else:
    return 'neutral'
```

```python
data['sentiment_label']=data['sentiment'].apply(classify_sentiment)
```
